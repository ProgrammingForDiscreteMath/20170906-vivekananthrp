# Learning by example

```python
def meru_prastaara(n):
    """
    Contruct the directed graph underlying Meru Prastaara up to ``n`` rows.
    """
    G = DiGraph()
    for i in range(n):
        G.add_edges([((i-j, j), (i-j+1, j)) for j in range(i+1)])
        G.add_edges([((i-j, j), (i-j, j+1)) for j in range(i+1)])
    return G
```

```python
M5 = meru_prastaara(5); M5.show()
```

```python
def meru_position_dictionary(n):
    """
    Return a dictionary of positions for nodes in the first ``n`` rows
    of Meru prastaara.
    
    The node ``(i,j)`` should be positioned at ``(-(i+j)/2+j,-i-j)``
    """
    d = dict()
    for i in range(n+1):
        for j in range(i+1):
            d[(i-j, j)] = (-i/2+j, -i)
    return d
```

```python
pos = meru_position_dictionary(5)
M5.show(pos=pos, vertex_labels=False)
```

```python
def meru_edge_colours(n):
    """
    Return the dictionary of edge colors for first ``n`` rows of 
    the Meru prastaara.
    """
    d = {"red":[], "blue":[]}
    for i in range(n):
        for j in range(i+1):
            if binomial(i, j)%2 == 1:
                if binomial(i+1, j)%2 == 1:
                    d["red"].append(((i-j,j), (i-j+1, j)))
                else:
                    d["blue"].append(((i-j,j), (i-j+1, j)))
                if binomial(i+i, j+1)%2 == 1:
                    d["red"].append(((i-j, j), (i-j, j+1)))
                else:
                    d["blue"].append(((i-j, j), (i-j, j+1)))
            else:
                d["blue"] += [((i-j, j), (i-j+1, j)), ((i-j,j),(i-j, j+1))]
    return d
```

```python
edge_colors=meru_edge_colours(5); edge_colors
```

```python
M5.show(pos=pos,edge_colors=edge_colors)
```

```python
M16 = Graph(meru_prastaara(16))
pos = meru_position_dictionary(16)
edge_colors = meru_edge_colours(16)
M16.show(vertex_labels=False, pos=pos, edge_colors=edge_colors, vertex_size=10)
```
